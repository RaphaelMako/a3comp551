```python
import sklearn.cluster
import numpy as np

# Load KMeans models
KMeans = sklearn.cluster.KMeans
MiniBatchKMeans = sklearn.cluster.MiniBatchKMeans

# Prepare data
data = Data.unlabeled_images
print(f'Unlabelled Image Data: {data.shape} ==> ', end='')
data = np.reshape(data, [-1, 3136])
data = data[:5000, :] # FOR TESTING, USE LESS OF THE DATASET
print(data.shape)

# Fit KMeans Clustering Model
k = 26 * 10 # One for each class
kmeans = KMeans(n_clusters=k, max_iter=1000)
kmeans = kmeans.fit(data)
centers = kmeans.cluster_centers_
```

```python
import Plotting
print(f'Centers shape: {centers.shape}')
Plotting.plot_images(centers, range(0, 10))

label_freqs = np.zeros([260])
print(label_freqs)
for label in kmeans.labels_:
    label_freqs[label] += 1

print(f'Label frequencies: {label_freqs}')

#print(kmeans.predict(np.reshape(Data.labelled_images, [-1, 3136])))
```
